## Anatomy of Matplotlib

```python
## Do some required imports
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x= [1, 4, 8, 15, 30]
y = [40, 20, 77, 60, 100]

# 2. Setup plot
fig, ax = plt.subplots(
    figsize=(7, 7)
)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(
    title="Simple plot",
    xlabel="x-axis",
    ylabel="y-axis",
)

# 5. Save and show (let's save the whole figure)
fig.savefig("saved_plots/sample-plot.png")
```

## Two options for subplots

```python
# subplot option one

fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(
    nrows=2,
    ncols=2,
    figsize=(12, 7)
)

# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.randint(0,10, (1,10)), np.random.randint(0,10, (1,10)));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(x);
```

```python
# subplot option two

fig, ax = plt.subplots(
    nrows = 2,
    ncols = 2,
    figsize=(12, 7)
)

# lets plot graphs with some data
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(x);
```

## Which one should we use Pyplot vs Matplotlib OO method ?

```python
heart_disease
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
## Pyplot method
over_50.plot(
    x="age",
    y="chol",
    kind="scatter",
    c="target"
);
```

```python
# OO method

fig, ax = plt.subplots(
    figsize=(10, 6)
)

over_50.plot(
    x="age",
    y="chol",
    kind="scatter",
    c="target",
    ax=ax
);

# ax.set_xlim([50, 80]);
```

```python
## OO method from stratch and better

fig, ax = plt.subplots(
    figsize=(10, 6)
)

scatter = ax.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target'],
);

ax.set(
    title="Heart disease and Chol levels",
    xlabel="Age",
    ylabel="Chol"
);

# lets add some legends
ax.legend(*scatter.legend_elements());

# lets create a horizontal line in middle
ax.axhline(y=over_50["chol"].mean(), color='r', linestyle='--')
```

```python
over_50.head()
```

```python
## Lets create a subplot with age, chol and thalach

fig, (ax0, ax1) = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(10, 10),
    sharex=True
);

# add data to ax0

scatter = ax0.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"],
);

ax0.set(
    ylabel="Cholestrol",
    title="Heart disease and cholestrol levels"
);

ax0.legend(
    *scatter.legend_elements(), title="Target"
)
ax0.axhline(color="r", linestyle="--", y=over_50["chol"].mean())

# now lets work on ax1

scatter1 = ax1.scatter(
    x=over_50["age"],
    y=over_50["thalach"],
    c=over_50["target"],
);

ax1.set(
    xlabel="Age",
    ylabel="Thalach",
    title="Heart Disease and Thalach levels"
);
ax1.legend(
    *scatter1.legend_elements(), title="Target"
)
ax1.axhline(
    color="green",
    linestyle="--",
    y=over_50["thalach"].mean()
);


# add a title to figure

fig.suptitle("Heart disease analysis", fontsize="16", fontweight="bold");
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

## OO method from stratch and better

fig, ax = plt.subplots(
    figsize=(10, 6)
)

scatter = ax.scatter(
    x=over_50['age'],
    y=over_50['chol'],
    c=over_50['target'],
    cmap="winter" # this will change the color
);

ax.set(
    title="Heart disease and Chol levels",
    xlabel="Age",
    ylabel="Chol"
);

# lets add some legends
ax.legend(*scatter.legend_elements());

# lets create a horizontal line in middle
ax.axhline(y=over_50["chol"].mean(), color='r', linestyle='--');
```

```python
## Lets customize the y and x axis labels

fig, (ax0, ax1) = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(10, 10),
    sharex=True
);

# add data to ax0

scatter = ax0.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"],
    cmap="winter",
);

ax0.set(
    ylabel="Cholestrol",
    title="Heart disease and cholestrol levels"
);

ax0.set_xlim([50, 80])
ax0.set_ylim([0, 600])

ax0.legend(
    *scatter.legend_elements(), title="Target"
)
ax0.axhline(color="r", linestyle="--", y=over_50["chol"].mean())

# now lets work on ax1

scatter1 = ax1.scatter(
    x=over_50["age"],
    y=over_50["thalach"],
    c=over_50["target"],
    cmap="winter",
);

ax1.set(
    xlabel="Age",
    ylabel="Thalach",
    title="Heart Disease and Thalach levels"
);

ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

ax1.legend(
    *scatter1.legend_elements(), title="Target"
)
ax1.axhline(
    color="green",
    linestyle="--",
    y=over_50["thalach"].mean()
);


# add a title to figure

fig.suptitle("Heart disease analysis", fontsize="16", fontweight="bold");
```
